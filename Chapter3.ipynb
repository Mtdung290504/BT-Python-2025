```python
import abc
from math import pi


class Shape(abc.ABC):
    @abc.abstractmethod
    def area(this):
        pass

    @abc.abstractmethod
    def perimeter(this):
        pass


class Rect(Shape):
    def __init__(this, width: float, height: float):
        this.width = width
        this.height = height

    def area(this):
        return this.width * this.height

    def perimeter(this):
        return 2 * (this.width + this.height)


class Square(Rect):
    def __init__(this, width: float):
        super().__init__(width, width)


class Circle(Shape):
    PI = pi

    def __init__(this, radius: float):
        this.radius = radius

    def area(this):
        return Circle.PI * this.radius**2

    def perimeter(this):
        return Circle.PI * pi * this.radius


r1 = Square(10)
print(r1.area(), r1.perimeter())

```
